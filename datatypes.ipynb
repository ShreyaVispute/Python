basic type - integer , float , boolean ,complex , string
container type - list , tuple , set , dictionary
user defined type - class

```python
#1. integer 
print(4)
print(1e308) #limit
print(1e309)
```

```python
#2. float
print(4.5)
print(1.7e308) #limit
print(1.7e309)
```

```python
#3. boolean
print(True)
print(False)
```

```python
#4. complex
print(3+7j)
print(9j+2)
```

```python
#5. string
print('Shreya')
print("Shreya")
print("""Shreya""")
```

```python
#container type data types
#1. list
print([1,2,3,4,5])
```

```python
#2. tuple
print((1,2,3,4,5))
```

```python
#3. set
print({1,2,3,4,4,5})
```
